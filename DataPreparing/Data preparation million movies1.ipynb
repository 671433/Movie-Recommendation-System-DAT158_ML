```python
df['title'] = df['title'].str.lower()
```

```python
df['genres'] = df['genres'].apply(lambda x: x.split(', ') if isinstance(x, str) else [])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

# TF-IDF setup
tfidf = TfidfVectorizer(analyzer='word')
tfidf_matrix = tfidf.fit_transform(df['genres'].apply(lambda x: ' '.join(x)))

# Movie similarity calculator
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)


def get_recommendations(title):

    idx = df.index[df['title'] == title].tolist()

    if not idx:
        return "Film not found."

    idx = idx[0]

    # Get similarity scores for other movies
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort movies by similarity score
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Take 10 similar movies
    sim_scores = sim_scores[1:11]

    # Get Similar Movie Titles
    movie_indices = [i[0] for i in sim_scores]

    return df['title'].iloc[movie_indices].tolist()
```

```python
print(get_recommendations('sexorcismes'))
```
