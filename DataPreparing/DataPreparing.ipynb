```python
import pandas as pd
import numpy as np

df = pd.read_csv('TMDB_all_movies.csv', engine='python')

df.head()
```

```python
# recommendation will be based on these features only
df = df.loc[:,['id','title', 'overview', 'popularity', 'tagline', 'genres','cast','director']]
```

```python
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
df.shape
```

```python
df.duplicated().sum()
```

```python
df = df.drop('popularity', axis=1)
```

```python
df.head()
```

```python
df.iloc[0]['overview']
```

```python
# handle overview (converting to list)
df['overview'] = df['overview'].apply(lambda x:x.split())
df.sample(5)
```

```python
# handle tagline (converting to list)
df['tagline'] = df['tagline'].apply(lambda x:x.split())
df.sample(5)
```

```python
# Handle genres (converting to list)
df['genres'] = df['genres'].apply(lambda x: [genre.strip() for genre in x.split(',')])
```

```python
df.sample(5)
```

```python
# Handle cast (converting to list)
df['cast'] = df['cast'].apply(lambda x: [cast.strip() for cast in x.split(',')])
df.sample(5)
```

```python
# Handle director (converting to list)
df['director'] = df['director'].apply(lambda x: [director.strip() for director in x.split(',')])
df.sample(5)
```

```python
# Removing space like that 'Alex Benjamin' 'AlexBenjamin'
def remove_space(anylist):
    return [i.replace(" ", "") for i in anylist]
```

```python
df['genres'] = df['genres'].apply(remove_space)
df['cast'] = df['cast'].apply(remove_space)
df['director'] = df['director'].apply(remove_space)
df.sample(2)
```

```python
# Concatinate all
df['concat'] = df['overview'] + df['tagline'] + df['genres'] + df['cast'] +  df['director'] 
df.sample(5)
```

```python
df.iloc[20]['concat']
```

```python
df_new = df[['id', 'title', 'concat']].copy()
```

```python
df_new.head()
```

```python
# Converting list to str
df_new['concat'] = df_new['concat'].apply(lambda x: " ".join(x))
df_new.head()
```

```python
df_new['concat'] = df_new['concat'].apply(lambda x: x.lower())  
df_new.iloc[10]['concat']
```

```python
# PorterStemmer is a specific stemming algorithm provided by NLTK, which reduces words to their base or root form (e.g., "running" to "run").
import nltk
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stems(text):
    T = []
    
    for i in text.split():
        T.append(ps.stem(i))
    
    return " ".join(T)
```

```python
df_new ['concat'] = df_new ['concat'].apply(stems)
```
