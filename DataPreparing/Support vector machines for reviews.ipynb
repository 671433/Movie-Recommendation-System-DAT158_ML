```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn import metrics
import pickle
from sklearn.preprocessing import LabelEncoder
from nltk.corpus import stopwords
import pandas as pd
import nltk 
```

```python
nltk.download("stopwords")
```

```python
ds = pd.read_csv('IMDB Dataset review.csv')
ds.head()
```

```python
X = ds['review']
y = ds['sentiment']

label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
```

```python
print("X values", X)
print("Sentiment labels after encoding:", y)
print("Classes:", label_encoder.classes_)
```

```python
stopSet = set(stopwords.words('english'))
stopSet = list(stopSet)
```

```python
vectorizer = TfidfVectorizer(use_idf=True, lowercase=True, strip_accents='ascii', stop_words=stopSet)
X = vectorizer.fit_transform(X)
```

```python
pickle.dump(vectorizer, open('review_vectorizer_SVM.pkl', 'wb'))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# Support vector machines model
svm_model = svm.SVC(kernel='linear') 
svm_model.fit(X_train, y_train)
```

```python
# Test
y_pred = svm_model.predict(X_test)
accuracy = metrics.accuracy_score(y_test, y_pred)
print(f"accuracy: {accuracy * 100:.2f}%")
```

```python
pickle.dump(svm_model, open('svm_sentiment_model.pkl', 'wb'))
```
