```python
import pandas as pd 
import numpy as np 
from sklearn.preprocessing import LabelEncoder
import nltk 
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn import naive_bayes
from sklearn.metrics import roc_auc_score, accuracy_score
import pickle
```

```python
nltk.download("stopwords")
```

```python
ds = pd.read_csv('IMDB Dataset review.csv')
```

```python
ds.head()
```

```python
ds.columns
```

```python
X = ds['review']
y = ds['sentiment']
```

```python
# convert string to number
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
```

```python
ds.head()
```

```python
stopSet = set(stopwords.words('english'))
```

```python
stopSet
```

```python
stopSet = list(stopSet)
```

```python
vectorizer = TfidfVectorizer(use_idf= True, lowercase= True, strip_accents= 'ascii',stop_words= stopSet)
```

```python
X = vectorizer.fit_transform(ds.review)
pickle.dump(vectorizer, open('review_vectorizer.pkl', 'wb'))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
nb_sentiment_model = naive_bayes.MultinomialNB()
nb_sentiment_model.fit(X_train, y_train)
```

```python
accuracy_score (y_test, nb_sentiment_model.predict(X_test))
```

```python
pickle.dump(nb_sentiment_model, open('nb_sentiment_model.pkl', 'wb'))
```
